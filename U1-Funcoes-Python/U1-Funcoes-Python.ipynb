```python
# Sistema de Gestão de Notas de Alunos

# Lista para armazenar as notas
notas = []

# Função para adicionar notas
def adicionar_notas():
    quantidade = int(input("Quantas notas deseja inserir? "))

    for i in range(quantidade):
        nota = float(input(f"Digite a nota {i + 1}: "))
        notas.append(nota)

# Função para calcular a média
def calcular_media(lista_notas):
    soma = sum(lista_notas)
    media = soma / len(lista_notas)
    return media

# Função para verificar situação do aluno
def verificar_situacao(media):
    if media >= 7:
        return "Aprovado"
    else:
        return "Reprovado"

# Função para exibir relatório final
def exibir_relatorio():
    print("\n===== RELATÓRIO FINAL =====")
    print(f"Notas inseridas: {notas}")

    media = calcular_media(notas)
    situacao = verificar_situacao(media)

    print(f"Média: {media:.2f}")
    print(f"Situação: {situacao}")

# Programa principal
adicionar_notas()
exibir_relatorio()
```
